```python
import matplotlib.pyplot as plt
```

```python
# Initialize an empty dictionary to hold the counts
counts = {}

# Open the file and read the data
with open('part-r-00000', 'r') as f:
    for line in f:
        # Each line should be in the format 'letter\tcount\n'
        letter, count = line.strip().split('\t')
        # Only consider lines where the key is a letter
        if letter.isalpha():
            counts[letter] = int(count)
```

```python

# Separate the letters and counts into two lists
letters = list(counts.keys())
frequencies = list(counts.values())
```

```python
# Create the bar plot
plt.figure(figsize=(10, 5))
plt.bar(letters, frequencies)
plt.xlabel('Letters')
plt.ylabel('Frequency')
plt.title('Frequency of Each Letter')
plt.show()
```
